```python
import math

import seaborn
import matplotlib.pyplot as plt
import numpy as np
import matplotlib

# matplotlib.use("pgf")
# matplotlib.rcParams.update({
#     "pgf.texsystem": "pdflatex",
#     'font.family': 'serif',
#     'text.usetex': True,
#     'pgf.rcfonts': False,
# })

t = np.linspace(0, 2*math.pi)
f1 = np.cos(t)
f2 = np.sin(t)

plt.plot(t, f1)
seaborn.lineplot(y=f2, x=t)
plt.grid()
plt.show()
# plt.savefig('histogram.pgf')
plt.close()
```
